## 4. Null values
Null values have negative influence on your dataframe. How much impact it has is only visable when you analyze
the dataframe against the your objective. 
With the **isnull()** you get for every entry with the value **true** if the the value of the cell is empty 
and **false** if the cell in the entry has an value.

With the **.sum()**  you count all the values within that column which has null values.

```python
df_food.isnull()
df_food.isnull().sum()
```

To drop null values use the **dropna()** to remove all null values out of your dataframe.

```python
df_food.dropna()
```

```python
dataframe.dropna(axis=1)
dataframe.tail(15)
```

## 5. Slice your dataframe
Not all columns has value for your objective. Columns that has no value for your model can easely removed 
from your dataframe.Remove columns from your dataframe called slicing (like a pie). 
In Python we slice with the open bracket and close braket and the column name. 

```python
coffee = df_food[['coffee','breakfast','Gender']]
coffee.head()
```

```python
coffee.describe()
```

## 6. Filtering your data
If we only wan't to see the data for men then we need to filter the column **Gender** equal to **men**. To do
this in python we use the == folowed by the filter you wan't to apply.

```python
coff = df_food[df_food['Gender'] == 1]
coff = coff[['coffee','breakfast','Gender']]
coff
```
